```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

* Threshold: 0 is black and 255 is white. Everything greater than 128 will turn to white, everything else is going to be black. 

```python
plt.figure(figsize=(22,16))

# Load the image
img = cv2.imread('coins1.jpg')

# Convert into RGB and grayscale image
img_coloured = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img_coloured, cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(img_coloured, 128, 255, cv2.THRESH_BINARY)
ret, thresh1 = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)



#Plot
plt.subplot(5,1,1)
plt.title('Coloured Image')
plt.xticks([])
plt.yticks([])
plt.imshow(img_coloured)

plt.subplot(5,1,2)
plt.title('Grayscale Image')
plt.xticks([])
plt.yticks([])
plt.imshow(gray, cmap='gray')

plt.subplot(5,1,3)
plt.title('Coloured Image - Threshold')
plt.xticks([])
plt.yticks([])
plt.imshow(thresh, cmap='gray')

plt.subplot(5,1,4)
plt.title('Grayscale Image - Threshold')
plt.xticks([])
plt.yticks([])
plt.imshow(thresh1, cmap='gray')

plt.subplot(5,1,5)
plt.title('Grayscale Image - Threshold - Otsu')
plt.xticks([])
plt.yticks([])
plt.imshow(thresh2, cmap='gray')


plt.show()
```
